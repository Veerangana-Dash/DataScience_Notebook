```python
#import packages for linear regression
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import seaborn as sns
sns.set()
```

```python
#loading csv data into dataframe called data
data = pd.read_csv('Simple_LR.csv')
```

```python
data
```

```python
#most useful descriptive statistics for dataframe
data.describe()
```

```python
y = data['GPA']
x1 = data['SAT']
```

```python
#plot points on graph. 
#Each point is a student
plt.scatter(x1,y)
plt.xlabel('SAT',fontsize=20)
plt.ylabel('GPA',fontsize=20)
plt.show
```

```python
#OLS Table summary
x = sm.add_constant(x1)
result = sm.OLS(y,x).fit()
result.summary()
```

```python
#Form Regression line
plt.scatter(x1,y)
#From OLS table : 
yhat = 0.0017*x1 + 0.275
fig = plt.plot(x1,yhat, lw=4, c = 'orange',label = 'regression line')
plt.xlabel('SAT',fontsize=20)
plt.ylabel('GPA',fontsize=20)
plt.show
```

```python
#null hypothesis if constant coefficient , ie, b0=0
plt.scatter(x1,y)
yhat = 0.0017*x1 + 0
fig = plt.plot(x1,yhat, lw=4, c = 'yellow',label = 'regression line')
plt.xlabel('SAT',fontsize=20)
plt.ylabel('GPA',fontsize=20)
plt.xlim(0)
plt.ylim(0)
plt.show
```

```python
#null hypothesis if quantifying coefficient , ie, b1=0
plt.scatter(x1,y)
yhat = 0*x1 + 0.275
fig = plt.plot(x1,yhat, lw=4, c = 'purple',label = 'regression line')
plt.xlabel('SAT',fontsize=20)
plt.ylabel('GPA',fontsize=20)
plt.show
```
