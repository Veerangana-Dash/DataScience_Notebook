## Preprocessing module :

#### Import libraries :

```python
import pandas as pd
```

#### Load data :

```python
raw_csv_data = pd.read_csv('Absenteeism_data.csv')
raw_csv_data.head()
```

#### Make a copy dataframe :

```python
df = raw_csv_data.copy()
df.head()
```

 #### Summary :

```python
df.info()
```

#### Drop columns :

```python
df = df.drop(['ID'], axis=1)
df_no_age = df.drop(['Age'], axis=1)
```

#### Dummies for categorical nominals :

```python
reason_columns = pd.get_dummies(df['Reason for Absence'])
reason_columns
```

```python
reason_columns['check'] = reason_columns.sum(axis=1)
reason_columns
```

```python
reason_columns = reason_columns.drop(['check'], axis =1)
```

```python
reason_columns = pd.get_dummies(df['Reason for Absence'],drop_first=True)
reason_columns.head()
```

```python
age_dummies = pd.get_dummies(df['Age'])
age_dummies.head()
```

#### Group features :

```python
df.columns.values
```

```python
reason_columns.columns.values
```

```python
df = df.drop(['Reason for Absence'],axis=1)
df.head()
```

```python
reason_columns.loc[:,1:14]
```

```python
reason_type_1 = reason_columns.loc[:,1:14].max(axis =1)
reason_type_2 = reason_columns.loc[:,15:17].max(axis =1)
reason_type_3 = reason_columns.loc[:,18:21].max(axis =1)
reason_type_4 = reason_columns.loc[:,22:].max(axis =1)
```

#### Concatinate groups :

```python
df = pd.concat([df,reason_type_1,reason_type_2, reason_type_3,reason_type_4],axis=1)
df.head()
```

```python
df_concatenated = pd.concat([df_no_age, age_dummies],axis=1)
df_concatenated.head()
```

#### Covert to datetime format :

```python
df_reason_mod['Date'] = pd.to_datetime(df_reason_mod['Date'])
df_reason_mod['Date']
```

```python
df_reason_mod['Date'] = pd.to_datetime(df_reason_mod['Date'], format = '%d/%m/%Y')
df_reason_mod['Date']
```

#### Extract month value :

```python
df_reason_mod['Date'][0]
```

```python
df_reason_mod['Date'][0].month
```

```python
list_months = []
list_months
```

```python
for i in range(df_reason_mod.shape[0]):
    list_months.append(df_reason_mod['Date'][i].month)
```

```python
list_months
```

```python
len(list_months)
```

```python
df_reason_mod['Month Value'] = list_months
df_reason_mod.head(20)
```

#### Extract the day of week : 

```python
#0:monday - 6:sunday
df_reason_mod['Date'][699].weekday()
```

```python
df_reason_mod['Date'][699]
```

```python
def date_to_weekday(date_value):
    return date_value.weekday()
```

```python
df_reason_mod['Day of the week'] = df_reason_mod['Date'].apply(date_to_weekday)
df_reason_mod.head()
```

```python
df_reason_mod = df_reason_mod.drop(['Date'],axis=1)
df_reason_mod.head()
```

```python
df_reason_mod.columns.values
```

```python
column_names =['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4','Month Value',
       'Day of the week','Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
```

```python
df_reason_mod = df_reason_mod[column_names]
df_reason_mod.head()
```

```python
df_reason_date_mod=df_reason_mod.copy()
df_reason_date_mod.head()
```
