# Minimal example with Tensorflow 2.0

### Import relevent libraries :

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

### Data generation :

```python
observations =1000

xs = np.random.uniform(low=-10,high=10,size=(observations,1))
zs = np.random.uniform(low=-10,high=10,size=(observations,1))

generated_inputs = np.column_stack((xs,zs))
noise = np.random.uniform(-1,1,(observations,1))
generated_targets = 2*xs -3*zs + 5 + noise

np.savez('TF_intro',inputs=generated_inputs,targets=generated_targets)
#saves .npz file in same directory
```

### Solving with TensorFlow :

```python
training_data = np.load('TF_intro.npz')
```

```python
input_size = 2
output_size = 1

#Model buidling:
model = tf.keras.Sequential([tf.keras.layers.Dense(output_size)])

#Objective function and optimisation algo:
model.compile(optimizer='sgd', loss='mean_squared_error' ) #configures model for training

#Train the data
model.fit(training_data['inputs'] ,training_data['targets'],epochs=100, verbose=0)
```
