# Simple Linear Regression

### Import libraries :

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

### Random input data for training :

```python
#Two variable problem with 1000 observations (2d matrix)
observations =1000

xs=np.random.uniform(low=-10,high=10,size=(observations,1))
zs=np.random.uniform(-10,10,(observations,1))

inputs = np.column_stack((xs,zs))

print (inputs.shape)
```

### Create targets :

```python
noise = np.random.uniform(-1,1,(observations,1))

targets = 2*xs -3*zs +5 +noise

print (targets.shape)
```

### Plot training data :

```python
#Plot is to see the string trend that the model should learn to reproduce 

targets = targets.reshape(observations,)
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')
ax.plot(xs,zs,targets)
ax.set_xlabel('xs')
ax.set_ylabel('ys')
ax.set_zlabel('zs')
ax.view_init(azim=100)
plt.show()
targets=targets.reshape(observations,1)
```

### Initialise variable :

```python
init_range =0.1

weights = np.random.uniform(-init_range, init_range, size=(2,1))

biases = np.random.uniform(-init_range, init_range, size=1)

print(weights)
print(biases)
```

### Learning Rate :

```python
learning_rate = 0.02
```

### Train the model :

```python
for i in range(100):
    outputs = np.dot(inputs,weights) + biases
    deltas = outputs - targets
    
    loss = np.sum(deltas **2) / 2 / observations
    
    print (loss)
    
    deltas_scaled = deltas / observations
    
    weights = weights - learning_rate * np.dot(inputs.T,deltas_scaled)
    biases = biases - learning_rate * np.sum(deltas_scaled)
```

#### Check weights and biases :

```python
print(weights,biases)
```

### Plot of optimal outputs vs targets :

```python
#The closer plot is to 45 degree line, the closer target and outout values:
plt.plot(outputs,targets)
plt.xlabel('outputs')
plt.ylabel('targets')
plt.show()
```
